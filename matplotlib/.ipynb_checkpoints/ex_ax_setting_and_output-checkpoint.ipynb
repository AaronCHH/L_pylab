## All levees

```python
fig, ax = plt.subplots(dpi=300)
# fig, ax = plt.subplots()
ax.plot(table2.index, table2['dem1'], 'k', 
        zorder=2)
ax.plot(table2.index, table2['dem6'], 'o', 
        markeredgecolor='k', 
        markerfacecolor='white', 
        markersize = 6,
        zorder=1)

ax.set_xlabel('Decending order of levee line vertices')
ax.set_ylabel('Elevation (m)')
ax.legend(['1m DEM', '6m DEM'], frameon=False)
ax.set_title('Levee Line Vertices Elevation')

fig.savefig('levee.png', bbox_inches='tight')
```

## YS Levee

```python
lvYS = gpd.read_file('94193088/shp_levee2model/lv3_YS_pt.shp')
```

```python
lvYS.columns
```

```python
table1 = lvYS[['dem1', 'dem6']]
```

```python
table2 = table1.sort_values(['dem1'], ascending=False)
```

```python
table2 = table2.reset_index(drop=True)
```

```python
table2.head()
```

```python
fig, ax = plt.subplots(dpi=300)
# fig, ax = plt.subplots()
ax.plot(table2.index, table2['dem1'], 'k', 
        zorder=2)
ax.plot(table2.index, table2['dem6'], 'o', 
        markeredgecolor='k', 
        markerfacecolor='white', 
        markersize = 6,
        zorder=1)

ax.set_xlabel('Decending order of levee line vertices')
ax.set_ylabel('Elevation (m)')
ax.set_ylim([4.8,8.2])
ax.set_yticks(np.arange(5,9))
ax.legend(['1m DEM', '6m DEM'], frameon=True)
ax.set_title('Levee Line Vertices Elevation')

fig.savefig('lvYS.png', bbox_inches='tight')
```
